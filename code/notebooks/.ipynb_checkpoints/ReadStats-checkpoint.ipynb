## Reading in read stats

```python
df = pd.read_csv(
    "../../data/DCIS_reg.2.txt",
    header=None,
    names=["cat", "sample", "reads"],
    sep="\t",
)
df = df.drop_duplicates()
df['sample'] = [s.split('/')[-1] for s in df['sample']]
df["sample_name"] = [s.split("_S")[0] for s in df["sample"]]
df['sample_name'] = [sample_alias[s] for s in df['sample_name']]
df['order'] = [s[-2:] for s in df['sample_name']]
df = df.sort_values(by=['order','sample_name'])
df = df.reset_index(drop=True)
df
```

```python
pct_usable = []
samples = []

for s in df["sample_name"].unique():
    sample_df = df[df["sample_name"] == s]
    total = int(sample_df[sample_df["cat"] == "total"]["reads"])
    pct_usable.append(int(sample_df[sample_df["cat"] == "adapter_trimmed"]["reads"]) / total)
    samples.append(s)
```

## Plotting UNALIGNED read categories

```python
colors = sns.color_palette("gist_ncar").as_hex()
unique_samples = df["sample_name"].unique()

usable_color = colors[0]
adapter_color = colors[4]

adapter_reads = df[df["cat"] == "adapter_trimmed"]["reads"].reset_index(
    drop=True
)

total_reads = df[df["cat"] == "total"]["reads"].reset_index(drop=True)

fig = stacked_bar(unique_samples,
            [adapter_reads,total_reads-adapter_reads],
            ['Usable','Adapter dimer'],
            [usable_color, adapter_color],
            xlabel="Sample",
            ylabel='# of reads'
           )
fig.show()
```

## Computing percentage of reads that are usable/contaminants/adapters etc

```python
pct_adapter = []
pct_usable = []

for s in unique_samples:
    sample_df = df[df["sample_name"] == s]

    if sample_df.shape[0] != 2:
        print("help")

    total_reads = int(sample_df[sample_df["cat"] == "total"]["reads"])
    adapter_reads = int(sample_df[sample_df["cat"] == "adapter_trimmed"]["reads"])

    pct_adapter.append((total_reads - adapter_reads) / float(total_reads))
    pct_usable.append((adapter_reads) / float(total_reads))
```

## Plotting UNALIGNED read categories by percentage

```python
fig = stacked_bar(df["sample_name"].unique(),
            [pct_usable, pct_adapter],
            ['Usable','Adapter dimer'],
            [usable_color, adapter_color],
            xlabel="Sample",
            ylabel='% of reads'
           )
fig.show()
```

# ALIGNED READS

```python
data_dct = {}
lst_ = []


for filename in glob.iglob("../../data/star_logs/*Log.final.out"):

    path = Path(filename)
    sample = path.stem.split("_")[0]

    with open(path) as f:
        for l in f:
            l = l.strip()
            if l and "|" in l:
                k, v = l.split("|")
                data_dct[k.strip()] = v.strip()

    lst_.append(
        pd.DataFrame.from_dict(data_dct, orient="index", columns=[sample])
    )
```

```python
map_df = pd.concat(lst_, join="outer", axis=1)

map_df.loc["Uniquely mapped reads %"] = [
    float(i[:-1]) for i in map_df.loc["Uniquely mapped reads %"]
]
map_df.loc["% of reads mapped to too many loci"] = [
    float(i[:-1]) for i in map_df.loc["% of reads mapped to too many loci"]
]
map_df.loc["% of reads unmapped: too short"] = [
    float(i[:-1]) for i in map_df.loc["% of reads unmapped: too short"]
]
map_df.loc["% of reads unmapped: other"] = [
    float(i[:-1]) for i in map_df.loc["% of reads unmapped: other"]
]

map_df.columns = [sample_alias[s] for s in map_df.columns]
map_df = map_df[df['sample_name'].unique()]
```

```python
fig = stacked_bar(list(map_df.columns),
            [map_df.loc["Uniquely mapped reads %"],
            map_df.loc["% of reads mapped to too many loci"],
            map_df.loc["% of reads unmapped: other"],
            map_df.loc["% of reads unmapped: too short"],],
            ["Uniquely mapped",
            "Multi mapped",
            "Other",
            "Too short"],
            ["#00ec00",
             "#056c39",
             "#e6e172",
             "#ffff33"],
            xlabel="Sample",
            ylabel='% of reads'
           )
fig.show()
```

# Feature assigned stats - AKA what fraction of reads are assigned to transcript

```python
filename= "../../data/counts/counts.txt.summary"

path = Path(filename)
sum_df = pd.read_csv(path, sep='\t', index_col=0)
sum_df.columns = [s.split('/')[-1].split('_S')[0] for s in sum_df.columns]

sum_df.columns = [sample_alias[s] for s in sum_df.columns]
sum_df = sum_df[df['sample_name'].unique()]
```

```python
assigned = sum_df.loc["Assigned"].reset_index(drop=True)
unassigned_no_f = sum_df.loc["Unassigned_NoFeatures"].reset_index(drop=True)
unassigned_ambig = sum_df.loc["Unassigned_Ambiguity"].reset_index(drop=True)
unassigned = unassigned_no_f + unassigned_ambig
total = assigned + unassigned
```

```python
as_col = sns.color_palette('PRGn').as_hex()
fig = stacked_bar(list(sum_df.columns),
            [assigned , unassigned],
            ['Assigned','Unassigned'],
            [as_col[1], as_col[0]],
            xlabel="Sample",
            ylabel='# reads'
           )
fig.show()
```

```python
as_col = sns.color_palette('PRGn').as_hex()
fig = stacked_bar(list(sum_df.columns),
            [assigned/total , unassigned/total],
            ['Assigned','Unassigned'],
            [as_col[1], as_col[0]],
            xlabel="Sample",
            ylabel='% reads'
           )
fig.show()
```

## Merging a few dataframes to generate larger dataframe for plotting all together

```python
sum_t = sum_df.transpose()
sum_t['sample_name'] = list(sum_t.index)

ss = pd.read_csv('../../data/sample_sheet.txt',sep='\t')
ss = ss.loc[:11]
ss = ss[ss.columns[:19]]
ss['sample_name'] = [sample_alias[s] for s in ss['sample_name']]

merged_df = sum_t.merge(ss,how='outer',on='sample_name')
merged_df = merged_df.loc[:11]

map_t = map_df.transpose()
map_t['sample_name'] = map_t.index
merged_df = merged_df.merge(map_t,how='outer',on='sample_name')

total_reads = df[df['cat'] == 'total']
merged_df = total_reads.merge(merged_df,on='sample_name',how='outer')
```

## Plot heatmap of Assigned to transcript/ total reads with some categorial values

```python
c_df = pd.DataFrame(data=(merged_df['Assigned']/merged_df['reads']))
c_df = c_df.transpose()
c_df.columns = merged_df['sample_name']
c_df.index = ['Assigned to transcript/ total reads']
c_df.loc['Fraction uniquely mapped reads'] = list(merged_df['Uniquely mapped reads %']/100)
del c_df.columns.name
c_df

col_names =  [('Sample: {}'.format(s),) for s in c_df.columns]

cats= ['BlockID', 'region comment',
       'year', 'area L,M,S', 'cellularity 1,2,3',
       'Dissection Q 1,2,3', 'Necrosis 0,1', 'SMART3-SEQ date', 
        'bead multiplexing',]

cat_dct = {}

for c in cats:
    cat_dct[c] = dict(zip(merged_df['sample_name'],merged_df[c]))

for i in range(len(col_names)):
    sample = merged_df['sample_name'][i]
    for c in cats:
        c_val = cat_dct[c][sample]
        col_names[i] = col_names[i] + ('{}:{}'.format(c,c_val),) 

c_df.columns = col_names
```

## Plotting total assigned reads per sample

```python
merged_df = merged_df.reset_index(drop=True)
sns.barplot(x='sample_name', 
            y='Assigned',
            data=merged_df,
            hue='BlockID',
            palette='Set2',
            dodge=False)
plt.xlabel('Samples',fontsize=9,fontweight='bold')
plt.ylabel('# of assigned reads',fontsize=9,fontweight='bold')
plt.xticks(rotation=45,ha='right')
for i in merged_df.index:
    cap_num = int(merged_df.loc[i]['bead multiplexing'])
    height = merged_df.loc[i]['Assigned']
    plt.text(i-0.1,height+1000,cap_num)
plt.legend(bbox_to_anchor=(1,0.7))
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x='reads', 
            y='Assigned',
            data=merged_df,
            hue='BlockID',
            palette='Set2',
            s=120,
            alpha=0.8,
            style='bead multiplexing')
plt.xlabel('Total reads',fontsize=9,fontweight='bold')
plt.ylabel('Assigned reads',fontsize=9,fontweight='bold')
plt.legend(bbox_to_anchor=(1.55,0.85))
```

```python
merged_df['assigned/total reads'] = merged_df['Assigned']/merged_df['reads']
sns.barplot(x='sample_name', 
            y='assigned/total reads',
            data=merged_df,
            hue='BlockID',
            palette='Set2',
            dodge=False)
plt.xlabel('Samples',fontsize=9,fontweight='bold')
plt.ylabel('assigned/total reads',fontsize=9,fontweight='bold')
plt.xticks(rotation=45,ha='right')
for i in merged_df.index:
    cap_num = int(merged_df['bead multiplexing'][i])
    height = merged_df['assigned/total reads'][i]
    plt.text(i-0.1,height+0.001,cap_num)
plt.ylim(0,0.13)
plt.legend(bbox_to_anchor=(1,0.7))
```
